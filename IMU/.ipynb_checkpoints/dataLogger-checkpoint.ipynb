## Datalogger conexion con esp32

```python
import serial
import pandas as pd

CANT_MUESTRAS = 4000     # Periodo de muestreo es 2ms, por lo que 0.002s * 4000 = 8segundos

ser = serial.Serial()
ser.baudrate = 115200
ser.port = 'COM8'
ser.open()

if(ser.is_open):
    print("Puerto serial conectado")
    
    samples = []
    linea = ser.readline()
    
    for i in range(CANT_MUESTRAS):
        linea = ser.readline()
        linea = str(linea)
        linea = linea.replace("\\r\\n","")
        linea = linea.replace("b'","")
        linea = linea.replace("'","")
        linea = linea.split(',')

        samples.append(linea)
        
if(len(samples) == CANT_MUESTRAS ):

    df = pd.DataFrame(samples) 
    header = ["ACC_X","ACC_Y","ACC_Z","GY_X","GY_Y","GY_Z"]              # Nombre de cada columna
    df.to_csv('muestras.csv',header = header,index = False) 

    print("Guardadas : " + str(len(samples)) + " muestras")
else:
    print("ERROR EN LA TOMA DE MUESTRAS\n")
ser.close()
```

# Objetivos:


##### El objetivo del presente trabajo final es obtener los angulos de posicion en el que se encuentra un sensor Kinect, para la posterior interpretacion de los datos proporcionados tanto por el sensor mencionado como por dichos angulos para el mapeo de superficies en interiores, con la finalidad de ser utilizado en un robot que sea capaz de navegar de manera autonoma.

##### Si bien el sensor Kinect ofrece un acelerometro de 3 ejes con un filtrado interno con el cual somos capaces de obtener los angulos X e Y, la falta de giroscopo hace que no sea posible obtener el angulo YAW del sensor, por lo que nos imposibilita utilizar dicho sensor para el objetivo propuesto, ya que se trata del angulo mas importante en la navegacion de un robot( siendo que el eje X e Y suelen estar paralelos al plano del suelo).

###### Por ello se decidio diseñar un modulo IMU que nos permita leer a intervalos regulares 3 ejes de un acelerometro y 3 ejes de un giroscopo en simultaneo, para asi hacer el tratamiento y filtrado de las señales obtenidas, con el fin de poder incorporar este modulo al sensor kinect, de manera que se muevan de manera solidaria, obteniendo asi por fin la posicion absoluta del kinect.


# Elementos utilizados:

##### Para la realizacion de este proyecto se utilizo un modulo MPU6050 de 6DOF, que incorpora tanto el giroscopo como el acelerometro y se comunica mediante protocolo i2c a un microcontrolador.

##### Un ESP32 programado en C para hacer la funcion de datalogger, es decir tomar muestras a intervalos regulares y enviarlos via puerto serial a la computadora, la cual posee un script de python que recibe los datos y los almacena en un .csv

## Lectura de datos a partir de .cvs

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
%matplotlib inline

plt.rcParams['figure.figsize'] = [10, 5]

SENSIBILITY_ACC = 16384.
SENSIBILITY_GY = 131.
RAD2DEG = 180#57.29577#9513082320876798154814105
PERIODO_MUESTREO = 0.002      # Frecuencia a la que esta muestreado los datos del MPU6050
FREC_MUESTREO = 1/PERIODO_MUESTREO

muestras = pd.read_csv("muestras.csv")
print("data shape: " + str(np.shape(muestras)))

ACC_X = muestras["ACC_X"][:]
ACC_Y = muestras["ACC_Y"][:]
ACC_Z = muestras["ACC_Z"][:]

ACC_X /= SENSIBILITY_ACC
ACC_Y /= SENSIBILITY_ACC
ACC_Z /= SENSIBILITY_ACC

lenMuestras = np.shape(muestras)[0]
plotX = np.linspace(0,PERIODO_MUESTREO * lenMuestras,lenMuestras)

plt.title("Acelerometro de 3 ejes crudo")
plt.xlabel("Tiempo[seg]")
plt.ylabel("Aceleracion[Gs]")
plt.plot(plotX,ACC_X,label = 'Raw X',color = 'tab:purple')
plt.plot(plotX,ACC_Y,label = 'Raw Y',color = 'tab:red')
plt.plot(plotX,ACC_Z,label = 'Raw Z',color = 'tab:cyan')
plt.legend()
plt.grid()
plt.show()

```

```python
GY_X = muestras["GY_X"][:]
GY_Y = muestras["GY_Y"][:]
GY_Z = muestras["GY_Z"][:]
GY_X /= SENSIBILITY_GY
GY_Y /= SENSIBILITY_GY
GY_Z /= SENSIBILITY_GY

plt.title("Giroscopo de 3 ejes crudo")
plt.xlabel("Tiempo[seg]")
plt.ylabel("Giro[rad/s]")
plt.plot(plotX,GY_X,label = 'Raw X',color = 'tab:purple')
plt.plot(plotX,GY_Y,label = 'Raw Y',color = 'tab:red')
plt.plot(plotX,GY_Z,label = 'Raw Z',color = 'tab:cyan')
plt.legend()
plt.grid()
plt.show()
```

#### Funcion para obtener respuesta en frecuencia y fase

```python
def getFreqResponse(coefFilter):

    w, h = signal.freqz(b=coefFilter)                   # agregar frecuencia de muestreo
    
    h_Phase = np.unwrap(np.arctan2(np.imag(h),np.real(h)))

    x = w * FREC_MUESTREO * 1.0 / (2 * np.pi)           # desnormalizado de frecuencia
    #y = 20 * np.log10(abs(h))                          # convierto amplitud a db
    y = np.absolute(h)
    return x,y,h_Phase
```

## Obtencion de angulos a partir del acelerometro mediante trigonometria

```python
angleX_ACC = np.arctan(ACC_X/np.sqrt(pow(ACC_Y,2) + pow(ACC_Z,2)))*(180.0/np.pi)
angleY_ACC = np.arctan(ACC_Y/np.sqrt(pow(ACC_X,2) + pow(ACC_Z,2)))*(180.0/np.pi)

plt.title("Angulos a partir de acelerometro mediante trigonometria")
plt.xlabel("Tiempo[seg]")
plt.ylabel("Angulos[°]")
plt.plot(plotX,angleX_ACC,label = 'Angle X ACC',color = 'tab:purple')
plt.plot(plotX,angleY_ACC,label = 'Angle Y ACC',color = 'tab:red')
plt.legend()
plt.grid()
plt.show()
```

## Obtencion de angulos a partir del giroscopo integrando la velocidad angular

```python
angleX_gyro = np.empty([4000], dtype = np.float64)
angleY_gyro = np.empty([4000], dtype = np.float64)
angleZ_gyro = np.empty([4000], dtype = np.float64)
angleX_gyro_calib = np.empty([4000], dtype = np.float64)
angleY_gyro_calib = np.empty([4000], dtype = np.float64)
angleZ_gyro_calib = np.empty([4000], dtype = np.float64)

# Obtengo el valor promedio de lecturas del giroscopo
promAngleGyX = sum(GY_Y[0:1000]) / len(GY_Y[0:1000])            ## SOLO UTILIZO LA PORCION DE SEÑAL DONDE ESTUVO QUIETA LA PLACA
promAngleGyY = sum(GY_X[0:1000]) / len(GY_X[0:1000])
promAngleGyZ = sum(GY_Z[0:500]) / len(GY_Z[0:500])
print("media calibration: X:" + str(promAngleGyX) + '     Y: ' + str(promAngleGyY)+ '     Z: ' + str(promAngleGyZ))

for i in range(1,4000):
    
    # PRESTAR ATENCION A POR QUE ESTAN INTERCAMBIADAS LAS POSICIONES X e Y
    angleX_gyro[i] = ( float(((GY_Y[i]/SENSIBILITY_GY) * PERIODO_MUESTREO) + angleX_gyro[i-1] ))
    angleY_gyro[i] = ( float(((GY_X[i]/SENSIBILITY_GY) * PERIODO_MUESTREO) + angleY_gyro[i-1] ))
    angleZ_gyro[i] = ( float(((GY_Z[i]/SENSIBILITY_GY) * PERIODO_MUESTREO) + angleZ_gyro[i-1] ))
    # Mismo calculo pero con los valores de giroscopo calibrados
    angleX_gyro_calib[i] = ( float((((GY_Y[i]- promAngleGyX)/SENSIBILITY_GY) * PERIODO_MUESTREO) + angleX_gyro_calib[i-1] ))
    angleY_gyro_calib[i] = ( float((((GY_X[i]- promAngleGyY)/SENSIBILITY_GY) * PERIODO_MUESTREO) + angleY_gyro_calib[i-1] ))
    angleZ_gyro_calib[i] = ( float((((GY_Z[i]- promAngleGyZ)/SENSIBILITY_GY) * PERIODO_MUESTREO) + angleZ_gyro_calib[i-1] ))
    
# Convierto de radianes a angulo.
angleX_gyro_deg = angleX_gyro * RAD2DEG * -1
angleY_gyro_deg = angleY_gyro * RAD2DEG
angleZ_gyro_deg = angleZ_gyro * RAD2DEG
angleX_gyro_deg_calib = angleX_gyro_calib * RAD2DEG *-1
angleY_gyro_deg_calib = angleY_gyro_calib * RAD2DEG
angleZ_gyro_deg_calib = angleZ_gyro_calib * RAD2DEG

plt.title("Angulos mediante integracion de velocidad angular") 
plt.xlabel("Tiempo[seg]")
plt.ylabel("Angulos[°]")
plt.plot(plotX,angleX_gyro_deg,label = 'Angle X GY without calib', alpha = 0.3, color = 'tab:purple')
plt.plot(plotX,angleY_gyro_deg,label = 'Angle Y GY without calib', alpha = 0.3, color = 'tab:red')
plt.plot(plotX,angleZ_gyro_deg,label = 'Angle Z GY without calib', alpha = 0.3, color = 'tab:green')
plt.plot(plotX,angleX_gyro_deg_calib,label = 'Angle X GY calib', color = 'tab:purple')
plt.plot(plotX,angleY_gyro_deg_calib,label = 'Angle Y GY calib', color = 'tab:red')
plt.plot(plotX,angleZ_gyro_deg_calib,label = 'Angle Z GY calib', color = 'tab:green')
plt.legend()
plt.grid()
plt.show()
```

# Filtrado de angulos del acelerometro mediante media movil

```python
def getMediaMovil(data,rondas):
    
    output = []
    for i in range(0,len(data)):
        sumatoria = 0
        for j in range( 0, rondas):
            if((i-j) >= 0):
                sumatoria +=  data[i - j]
        output.append(sumatoria/rondas)
    return output

def getCoefficients(N):
    coefFilter = np.ones(N) 
    coefFilter /= N
    
    return coefFilter

angleX_media_ACC_5 = getMediaMovil(angleX_ACC,5)
angleY_media_ACC_5 = getMediaMovil(angleY_ACC,5)

print("Delay filtro media movil de 5 muestras: " + str(5 * PERIODO_MUESTREO*1000) + "ms")

angleX_media_ACC_25 = getMediaMovil(angleX_ACC,25)
angleY_media_ACC_25 = getMediaMovil(angleY_ACC,25)

print("Delay filtro media movil de 25 muestras: " + str(25 * PERIODO_MUESTREO*1000) + "ms")

angleX_media_ACC_50 = getMediaMovil(angleX_ACC,50)
angleY_media_ACC_50 = getMediaMovil(angleY_ACC,50)

print("Delay filtro media movil de 50 muestras: " + str(50 * PERIODO_MUESTREO*1000) + "ms")

fig1, (plotM5,freqZ1,phase1,plotM25,freqZ2,phase2,plotM50,freqZ3,phase3) = plt.subplots(nrows=9,figsize = (15,30), layout="constrained")

fig1.suptitle("Angulos a partir de acelerometro con media Movil")

plotM5.set_title("Media Movil de 5 Muestras")
plotM5.set_xlabel("Tiempo[seg]")
plotM5.set_ylabel("Angulo[°]")
plotM5.grid()
plotM5.plot(plotX,angleX_media_ACC_5, label = 'Angle X - media ACC', color = 'tab:purple')
plotM5.plot(plotX,angleY_media_ACC_5, label = 'Angle Y - media ACC', color = 'tab:red')
plotM5.plot(plotX,angleX_ACC, label = 'Angle X - ACC RAW', color = 'tab:purple' , alpha = 0.3)
plotM5.plot(plotX,angleY_ACC, label = 'Angle Y - ACC RAW', color = 'tab:red' , alpha = 0.3)
plotM5.legend(loc="upper left")

x,y,phasey = getFreqResponse(getCoefficients(5))

freqZ1.set_title("Respuesta en frecuencia media movil de 5 Muestras")
freqZ1.set_xlabel("Frecuencia [hz]")
freqZ1.set_ylabel("Amplitud[db]")
freqZ1.grid()
freqZ1.plot(x,y, color = 'tab:orange')

phase1.set_title("Fase media movil de 5 Muestras")
phase1.set_xlabel("Frecuencia [hz]")
phase1.set_ylabel("Fase[rad]")
phase1.grid()
phase1.plot(x,phasey, color = 'tab:cyan')

plotM25.set_title("Media Movil de 25 Muestras")
plotM25.set_xlabel("Tiempo[seg]")
plotM25.set_ylabel("Angulo[°]")
plotM25.grid()
plotM25.plot(plotX,angleX_media_ACC_25, label = 'Angle X - media ACC' ,color = 'tab:purple')
plotM25.plot(plotX,angleY_media_ACC_25, label = 'Angle Y - media ACC' ,color = 'tab:red')
plotM25.plot(plotX,angleX_ACC, label = 'Angle X - ACC RAW', color = 'tab:purple' , alpha = 0.3)
plotM25.plot(plotX,angleY_ACC, label = 'Angle Y - ACC RAW', color = 'tab:red' , alpha = 0.3)
plotM25.legend(loc="upper left")

x,y,phasey = getFreqResponse(getCoefficients(25))

freqZ2.set_title("Respuesta en frecuencia media movil de 25 Muestras")
freqZ2.set_xlabel("Frecuencia [hz]")
freqZ2.set_ylabel("Amplitud[db]")
freqZ2.grid()
freqZ2.plot(x,y, color = 'tab:orange')

phase2.set_title("Fase media movil de 25 Muestras")
phase2.set_xlabel("Frecuencia [hz]")
phase2.set_ylabel("Fase[rad]")
phase2.grid()
phase2.plot(x,phasey, color = 'tab:cyan')

plotM50.set_title("Media Movil de 50 Muestras")
plotM50.set_xlabel("Tiempo[seg]")
plotM50.set_ylabel("Angulo[°]")
plotM50.grid()
plotM50.plot(plotX,angleX_media_ACC_50, label = 'Angle X - media ACC' ,color = 'tab:purple')
plotM50.plot(plotX,angleY_media_ACC_50, label = 'Angle Y - media ACC' ,color = 'tab:red')
plotM50.plot(plotX,angleX_ACC, label = 'Angle X - ACC RAW', color = 'tab:purple' , alpha = 0.3)
plotM50.plot(plotX,angleY_ACC, label = 'Angle Y - ACC RAW', color = 'tab:red' , alpha = 0.3)
plotM50.legend(loc="upper left")

x,y,phasey = getFreqResponse(getCoefficients(50))

freqZ3.set_title("Respuesta en frecuencia media movil de 50 Muestras")
freqZ3.set_xlabel("Frecuencia [hz]")
freqZ3.set_ylabel("Amplitud[db]")
freqZ3.grid()
freqZ3.plot(x,y, color = 'tab:orange')

phase3.set_title("Fase media movil de 50 Muestras")
phase3.set_xlabel("Frecuencia [hz]")
phase3.set_ylabel("Fase[rad]")
phase3.grid()
phase3.plot(x,phasey, color = 'tab:cyan')

plt.show()
```

# Filtrado de angulos del acelerometro mediante un filtro FIR pasa-bajos

```python
M= 101                                                             # Orden del filtro, COMO LO DETERMINO ???

frecCorte = 10                                                     # frecuencia de corte en hertz

nyq_rate = FREC_MUESTREO / 2.0                                     # The Nyquist rate of the signal.

b = signal.firwin(M, frecCorte/nyq_rate, window=('hamming'),pass_zero = "lowpass")# ventana de Hamming de N orden,me devuelve los coeficientes del filtro

print("Delay del filtro FIR: " + str( (M - 1) / (2 * FREC_MUESTREO) * 1000) + "ms")

angleX_FIR_INPUT_FFT = np.abs(np.fft.fft(angleX_ACC))              # FFT de la entrada del filtro, anguloX del acc
angleY_FIR_INPUT_FFT = np.abs(np.fft.fft(angleY_ACC))              # FFT de la entrada del filtro, anguloY del acc

angleX_ACC_filtered = signal.lfilter(b, 1.0, angleX_ACC)           # aplico el filtro a la señal cruda
angleY_ACC_filtered = signal.lfilter(b, 1.0, angleY_ACC)           # aplico el filtro a la señal cruda

angleX_FIR_OUTPUT_FFT = np.abs(np.fft.fft(angleX_ACC_filtered))    # FFT de la salida del filtro, anguloX del acc
angleY_FIR_OUTPUT_FFT = np.abs(np.fft.fft(angleY_ACC_filtered))    # FFT de la salida del filtro, anguloY del acc

fig1, (plot1,fftX, fftY,freqZ,phase) = plt.subplots(nrows=5,figsize = (15,25), layout="constrained")
fig1.suptitle("Angulos del acelerometro filtrado por FIR pasa-bajos")

plot1.set_title("Señal filtrada en el dominio del tiempo")
plot1.set_xlabel("Tiempo[seg]")
plot1.set_ylabel("Angulo[°]")
plot1.grid()
plot1.plot(plotX,angleX_ACC_filtered, label = 'Angle X - FIR', color = 'tab:purple')
plot1.plot(plotX,angleY_ACC_filtered, label = 'Angle Y - FIR', color = 'tab:red')
plot1.plot(plotX,angleX_ACC, label = 'Angle X - ACC RAW', color = 'tab:purple' , alpha = 0.3)
plot1.plot(plotX,angleY_ACC, label = 'Angle Y - ACC RAW', color = 'tab:red' , alpha = 0.3)
plot1.legend(loc="upper left")

plotXFrec = np.linspace(0,FREC_MUESTREO/2,len(angleX_ACC))

fftX.set_title("Transformada de fourier del angleX ACC")
fftX.set_xlabel("Frecuencia[hz]")
fftX.set_ylabel("Amplitud[db]")
fftX.grid()
fftX.plot(plotXFrec[:400],angleX_FIR_OUTPUT_FFT[:400], label = 'Angle X - FIR', color = 'tab:purple')
fftX.plot(plotXFrec[:400],angleX_FIR_INPUT_FFT[:400], label = 'Angle X - ACC RAW', color = 'tab:purple' , alpha = 0.3)
fftX.legend(loc="upper left")


fftY.set_title("Transformada de fourier del angleY ACC")
fftY.set_xlabel("Frecuencia[hz]")
fftY.set_ylabel("Amplitud[db]")
fftY.grid()
fftY.plot(plotXFrec[:400],angleY_FIR_OUTPUT_FFT[:400], label = 'Angle Y - FIR', color = 'tab:red')
fftY.plot(plotXFrec[:400],angleY_FIR_INPUT_FFT[:400], label = 'Angle Y - ACC RAW', color = 'tab:red' , alpha = 0.3)
fftY.legend(loc="upper left")

x,y,phasey = getFreqResponse(b)

freqZ.set_title("Respuesta en frecuencia filtro FIR pasa-bajos")
freqZ.set_xlabel("Frecuencia [hz]")
freqZ.set_ylabel("Amplitud[db]")
freqZ.grid()
freqZ.plot(x,y, color = 'tab:orange')


phase.set_title("Fase filtro FIR pasa-bajos")
phase.set_xlabel("Frecuencia [hz]")
phase.set_ylabel("Fase[rad]")
phase.grid()
phase.plot(x,phasey, color = 'tab:cyan')

plt.show()
```

# Filtrado de angulos del giroscopo mediante un filtro FIR pasa-altos

```python
M= 51                                                               # Orden del filtro, COMO LO DETERMINO ???

frecCorte = 2                                                       # frecuencia de corte en hertz

nyq_rate = FREC_MUESTREO / 2.0                                      # The Nyquist rate of the signal.

b = signal.firwin(M, frecCorte/nyq_rate, window=('hamming'),pass_zero = "highpass") # ventana de Hamming de N orden,me devuelve los coeficientes del filtro
b = -b                                                              # inversion espectral

angleX_FIR_INPUT_FFT = np.abs(np.fft.fft(angleX_gyro_deg_calib))    # FFT de la entrada del filtro, anguloX del gy
angleY_FIR_INPUT_FFT = np.abs(np.fft.fft(angleY_gyro_deg_calib))    # FFT de la entrada del filtro, anguloY del gy
angleZ_FIR_INPUT_FFT = np.abs(np.fft.fft(angleZ_gyro_deg_calib))    # FFT de la entrada del filtro, anguloZ del gy

angleX_GY_filtered = -signal.lfilter(b, 1.0, angleX_gyro_deg_calib) # aplico el filtro a la señal cruda
angleY_GY_filtered = -signal.lfilter(b, 1.0, angleY_gyro_deg_calib) # aplico el filtro a la señal cruda
angleZ_GY_filtered = -signal.lfilter(b, 1.0, angleZ_gyro_deg_calib) # aplico el filtro a la señal cruda

angleX_FIR_OUTPUT_FFT = np.abs(np.fft.fft(angleX_GY_filtered))      # FFT de la salida del filtro, anguloX del gy
angleY_FIR_OUTPUT_FFT = np.abs(np.fft.fft(angleY_GY_filtered))      # FFT de la salida del filtro, anguloY del gy
angleZ_FIR_OUTPUT_FFT = np.abs(np.fft.fft(angleZ_GY_filtered))      # FFT de la salida del filtro, anguloZ del gy

fig1, (plot1,fftX, fftY, fftZ,freqZ,phase) = plt.subplots(nrows=6,figsize = (15,25), layout="constrained")
fig1.suptitle("Angulos del giroscopo filtrado por FIR pasa-altos")

plot1.set_title("Señal filtrada en el dominio del tiempo")
plot1.set_xlabel("Tiempo[seg]")
plot1.set_ylabel("Angulo[°]")
plot1.grid()
plot1.plot(plotX,angleX_GY_filtered, label = 'Angle X - FIR', color = 'tab:purple')
plot1.plot(plotX,angleY_GY_filtered, label = 'Angle Y - FIR', color = 'tab:red')
plot1.plot(plotX,angleZ_GY_filtered, label = 'Angle Z - FIR', color = 'tab:green')
plot1.plot(plotX,angleX_gyro_deg_calib, label = 'Angle X - GY RAW', color = 'tab:purple' , alpha = 0.3)
plot1.plot(plotX,angleY_gyro_deg_calib, label = 'Angle Y - GY RAW', color = 'tab:red' , alpha = 0.3)
plot1.plot(plotX,angleZ_gyro_deg_calib, label = 'Angle Z - GY RAW', color = 'tab:green' , alpha = 0.3)
plot1.legend(loc="upper left")

plotXFrec = np.linspace(0,FREC_MUESTREO/2,len(angleX_ACC))

fftX.set_title("Transformada de fourier del angleX GY")
fftX.set_xlabel("Frecuencia[hz]")
fftX.set_ylabel("Amplitud[db]")
fftX.grid()
fftX.plot(plotXFrec[:400],angleX_FIR_OUTPUT_FFT[:400], label = 'Angle X - FIR', color = 'tab:purple')
fftX.plot(plotXFrec[:400],angleX_FIR_INPUT_FFT[:400], label = 'Angle X - GY RAW', color = 'tab:purple' , alpha = 0.3)
fftX.legend(loc="upper left")


fftY.set_title("Transformada de fourier del angleY GY")
fftY.set_xlabel("Frecuencia[hz]")
fftY.set_ylabel("Amplitud[db]")
fftY.grid()
fftY.plot(plotXFrec[:400],angleY_FIR_OUTPUT_FFT[:400], label = 'Angle Y - FIR', color = 'tab:red')
fftY.plot(plotXFrec[:400],angleY_FIR_INPUT_FFT[:400], label = 'Angle Y - GY RAW', color = 'tab:red' , alpha = 0.3)
fftY.legend(loc="upper left")


fftZ.set_title("Transformada de fourier del angleY GY")
fftZ.set_xlabel("Frecuencia[hz]")
fftZ.set_ylabel("Amplitud[db]")
fftZ.grid()
fftZ.plot(plotXFrec[:400],angleY_FIR_OUTPUT_FFT[:400], label = 'Angle Y - FIR', color = 'tab:green')
fftZ.plot(plotXFrec[:400],angleY_FIR_INPUT_FFT[:400], label = 'Angle Y - GY RAW', color = 'tab:green' , alpha = 0.3)
fftZ.legend(loc="upper left")

x,y,phasey = getFreqResponse(b)

freqZ.set_title("Respuesta en frecuencia filtro FIR pasa-altos")
freqZ.set_xlabel("Frecuencia [hz]")
freqZ.set_ylabel("Amplitud[db]")
freqZ.grid()
freqZ.plot(x,y, color = 'tab:orange')


phase.set_title("Fase filtro FIR pasa-altos")
phase.set_xlabel("Frecuencia [hz]")
phase.set_ylabel("Fase[rad]")
phase.grid()
phase.plot(x,phasey, color = 'tab:cyan')

plt.show()
```

# Filtrado de angulos mediante filtro complementario

```python
PONDERACION_ACC = 0.02
PONDERACION_GY = 0.98

angleX_compl = np.empty([4000], dtype = np.float64)
angleY_compl = np.empty([4000], dtype = np.float64)

for i in range(1,4000):
#     Filtro complementario:   AnguloAct = PONDERACION_GY * (AnguloAnt + wGyro*dt) + (PONDERACION_ACC * AnguloACC)
    angleX_compl[i] = (PONDERACION_GY*(angleX_compl[i-1]+(GY_Y[i]/SENSIBILITY_GY)* PERIODO_MUESTREO)) + (PONDERACION_ACC*ACC_X[i]) * 90
    angleY_compl[i] = (PONDERACION_GY*(angleY_compl[i-1]+(GY_X[i]/SENSIBILITY_GY)* PERIODO_MUESTREO)) + (PONDERACION_ACC*ACC_Y[i]) * 90

fig1, (plot1,plot2, fftX, fftY) = plt.subplots(nrows=4,figsize = (15,20), layout="constrained")

fig1.suptitle("Angulos mediante filtro complementario")

plot1.set_title("Filtro complementario vs angle ACC")
plot1.set_xlabel("Tiempo[seg]")
plot1.set_ylabel("Angulo[°]")
plot1.grid()
plot1.plot(plotX,angleX_compl, label = 'Angle X - Complementario', color = 'tab:purple')
plot1.plot(plotX,angleY_compl, label = 'Angle Y - Complementario', color = 'tab:red')
plot1.plot(plotX,angleX_ACC, label = 'Angle X - ACC RAW', color = 'tab:purple' , alpha = 0.3)
plot1.plot(plotX,angleY_ACC, label = 'Angle Y - ACC RAW', color = 'tab:red' , alpha = 0.3)
plot1.legend(loc="upper left")

plot2.set_title("Filtro complementario vs GYRO calib")
plot2.set_xlabel("Tiempo[seg]")
plot2.set_ylabel("Angulo[°]")
plot2.grid()
plot2.plot(plotX,angleX_compl, label = 'Angle X - Complementario' ,color = 'tab:purple')
plot2.plot(plotX,angleY_compl, label = 'Angle Y - Complementario' ,color = 'tab:red')
plot2.plot(plotX,angleX_gyro_deg_calib, label = 'Angle X - GY calib', color = 'tab:purple' , alpha = 0.3)
plot2.plot(plotX,angleY_gyro_deg_calib, label = 'Angle Y - GY calib', color = 'tab:red' , alpha = 0.3)
plot2.legend(loc="upper left")

angleX_COMPL_OUTPUT_FFT = np.abs(np.fft.fft(angleX_compl))
angleY_COMPL_OUTPUT_FFT = np.abs(np.fft.fft(angleY_compl))

fftX.set_title("Transformada de fourier del angleX COMPL")
fftX.set_xlabel("Frecuencia[hz]")
fftX.set_ylabel("Amplitud[db]")
fftX.grid()
fftX.plot(plotXFrec[:400],angleX_COMPL_OUTPUT_FFT[:400], label = 'Angle X - COMPL', color = 'tab:purple')
#fftX.plot(plotXFrec[:400],angleY_FIR_INPUT_FFT[:400], label = 'Angle Y - ACC RAW', color = 'tab:purple' , alpha = 0.3)
fftX.legend(loc="upper left")

fftY.set_title("Transformada de fourier del angleY COMPL")
fftY.set_xlabel("Frecuencia[hz]")
fftY.set_ylabel("Amplitud[db]")
fftY.grid()
fftY.plot(plotXFrec[:400],angleY_COMPL_OUTPUT_FFT[:400], label = 'Angle Y - COMPL', color = 'tab:red')
#fftY.plot(plotXFrec[:400],angleY_FIR_INPUT_FFT[:400], label = 'Angle Y - GY RAW', color = 'tab:red' , alpha = 0.3)
fftY.legend(loc="upper left")
plt.show()
```

## Conclusión

 - Como conclusión podemos empezar destacando la diferencia entre un sensor acelerómetro y un giróscopo, por un lado vemos que la señal del acelerómetro es muy ruidosa, y no responde bien a las altas frecuencias(que en nuestro caso representa movimientos bruscos y rápidos del sensor kinect), obligandonos a filtrar para poder utilizarlo, por otro lado el giroscopio ofrece una señal bastante limpia en término de ruidos de alta frecuencia, pero se obtiene una deriva que es realmente compleja de corregir.
 
- Deriva del giroscopio:  se realizó un primer intento corrigiendo el offset detectado cuando el sensor estaba quieto y completamente horizontal (ver gráfico "Obtención de ángulos a partir del giroscopio integrando la velocidad angular"), allí se puede apreciar que la corrección de este offset contribuye a compensar dicha deriva, pero cuando la placa se empieza a mover, esa deriva inevitablemente termina utilizando la señal obtenida.
 
 
 - Como un primer intento de filtrar la señal del acelerómetro se utilizó un filtro de media movil de distintos ordenes, alli se puede ver que se es capaz de filtrar la señal hasta lograr un resultado bastante aceptable en términos de rechazo de ruido, a costa de un alto delay de la salida del filtro,por lo que nos encontramos en una relación de compromiso entre calidad de filtrado vs Delay.
 
 
 - Como segundo intento se aplicó un filtro FIR pasa-bajo con una frecuencia de corte de 10hz y de orden M=101, este nos permite tener un mayor control respecto a la frecuencia de corte a partir de la cual deseamos atenuar, a la vez que nos permite poder diseñar un filtro específico en función de la cantidad de muestras,el tipo de ventana(utilizada "hamming" para nuestro ejemplo), etc. El resultado de este filtro depende del orden y la frecuencia de corte, al igual que en el caso de la media móvil, cuando mayor es el orden del filtro, mayor es el delay en la salida, aun asi se adecua bastante a los resultados esperados para la aplicación de la señal.
 
 
 - Se intentó además utilizar un filtro FIR pasa-alto para la señal del giroscopio, con la intención de eliminar la deriva producida por la señal de muy baja frecuencia, allí nos encontramos con que la pérdida de información producto del filtro es demasiado grande, lo que nos termina atenuando la señal,dejándola inutilizable, la frecuencia de corte utilizada para este filtro fue de 2 Hz, lo cual es demasiado, pero incluso bajando aún más, se es incapaz de obtener una respuesta aceptable, otro de los cambios que se ensayaron fue utilizar una ventana cuadrada, lo cual permite un menor ancho de la banda de transición del filtro, para así hacer una atenuación "más limpia" a partir de la frecuencia de corte, aun así este tipo de solución no resulta viable, ya que es imposible encontrar una frecuencia de corte que nos permita discernir entre la deriva y un movimiento real de muy baja frecuencia.
 
 
 
 - El último filtro implementado se trata de un filtro complementario, el cual "mezcla" ambas señales multiplicadas por un factor de ponderación, para así aprovechar las ventajas de cada sensor y compensar sus falencias, se puede apreciar allí una señal más limpia, con menores componentes de alta frecuencia producto del ruido, y una amplitud que en ocasiones supera a la señal del acelerómetro y/o giroscopio de manera individual.
